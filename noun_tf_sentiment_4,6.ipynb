```python
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


def preprocess_text(text):
    # Remove HTML tags (if any)
    text = re.sub(r"<.*?>", "", text)
    
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text)
    
    # Convert to lowercase
    tokens = [token.lower() for token in tokens]
    
    # Remove punctuation
    tokens = [token for token in tokens if token.isalpha()]
    
    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    # print(filtered_tokens)
    # # Lemmatization
    # lemmatizer = WordNetLemmatizer()
    # lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a single string
    processed_text = " ".join(filtered_tokens)

    
    return processed_text
    # return filtered_tokens 
```

```python

from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification

MODEL = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = TFAutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
def polarity_scores_roberta(example, tokenizer, model):
    # Encode the text for TensorFlow
    encoded_text = tokenizer(example, return_tensors='tf')

    # Pass the encoded text to the model
    output = model(encoded_text)

    # Extract the logits and apply softmax
    scores = output[0][0].numpy()


    # scores = output[0].numpy()
    scores = softmax(scores)

    # Create a dictionary of scores
    scores_dict = {
        'roberta_neg': scores[0],
        'roberta_neu': scores[1],
        'roberta_pos': scores[2]
    }
    return scores_dict
```

```python
res = {}
for myid in tqdm(df.index):
    text = df.loc[myid, 'Reviews']

    try:

        # RoBERTa Sentiment Analysis
        roberta_result = polarity_scores_roberta(text, tokenizer, model)  # Ensure tokenizer and model are defined

        res[myid] = roberta_result

    except Exception as e:
        print(f'Broke for id {myid}: {str(e)}')
```

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Sample dataset
data = {
    'Feature1': [1.1, 2.3, 3.3, 4.4, 5.5],
    'Feature2': [2.1, 3.4, 4.5, 5.6, 6.7],
    'Category': ['A', 'B', 'A', 'B', 'C']
}
df = pd.DataFrame(data)

# Standardizing the data (excluding the categorical column)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Feature1', 'Feature2']])

# Apply Gaussian Mixture Model
gmm = GaussianMixture(n_components=2, random_state=0)
df['Cluster'] = gmm.fit_predict(scaled_features)

# Get unique values and clusters
unique_values = df['Category'].unique()
clusters = df['Cluster'].unique()

# Create a DataFrame to map each unique value to its clusters
feature_cluster_table = pd.DataFrame(index=unique_values, columns=clusters)

for value in unique_values:
    for cluster in clusters:
        # Check if the value is in this cluster
        in_cluster = df[(df['Category'] == value) & (df['Cluster'] == cluster)].shape[0] > 0
        feature_cluster_table.at[value, cluster] = in_cluster

# Replace True/False with Yes/No (or 1/0) for better readability if desired
feature_cluster_table = feature_cluster_table.replace({True: 'Yes', False: 'No'})

# Display the table
print(feature_cluster_table)
```

```python
for value in unique_values:
    for cluster in clusters:
        # Count occurrences of the value in this cluster
        count = df[(df['Category'] == value) & (df['Cluster'] == cluster)].shape[0]
        feature_cluster_table.at[value, cluster] = count

# Display the table
print(feature_cluster_table)
```
